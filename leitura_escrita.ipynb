```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql import *
from datetime import datetime
```

```python
spark = SparkSession.builder.getOrCreate()
```

```python
df = spark.read.csv('videos-stats.csv')
```

```python
df.show(8)
```

```python
df.printSchema()
```

```python
df = spark.read.option('header', 'true').option('inferSchema', 'true').csv('videos-stats.csv')
df.printSchema()
```

```python
df.write.option('header', 'true').save('videos-parquet')
```

```python
df = spark.read.option('header', 'true').parquet('videos-parquet')
df.show()
```

```python
df.write.option('header', 'true').saveAsTable("tb_videos")
```

```python
spark.catalog.listTables()
```

```python
tab_sql_df = spark.sql("select * from tb_videos")
tab_sql_df.show()
```
